```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

def extract_hotel_data(hotel_url):
    from selenium import webdriver
    from selenium.webdriver.chrome.service import Service
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    from webdriver_manager.chrome import ChromeDriverManager
    import time

    options = webdriver.ChromeOptions()
    options.add_argument('--start-maximized')
    options.add_argument('--disable-blink-features=AutomationControlled')
    
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    wait = WebDriverWait(driver, 15)
    data = []

    try:
        driver.get(hotel_url)
        time.sleep(4)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 2);")
        time.sleep(3)

        hotel_name = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '[data-testid="title"]'))).text
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '[data-testid="room-details-card"]')))
        room_blocks = driver.find_elements(By.CSS_SELECTOR, '[data-testid="room-details-card"]')

        for room in room_blocks:
            try:
                room_type = room.find_element(By.CSS_SELECTOR, '[data-testid="room-name"]').text
            except:
                room_type = "N/A"

            try:
                price = room.find_element(By.CSS_SELECTOR, '[data-testid="price-and-discounted-price"]').text
            except:
                price = "N/A"

            try:
                breakfast = room.find_element(By.XPATH, './/div[contains(text(), "breakfast")]').text
            except:
                breakfast = "Not mentioned"

            data.append({
                'Hotel': hotel_name,
                'Room Type': room_type,
                'Price': price,
                'Breakfast': breakfast
            })

    except Exception as e:
        print(f"Error scraping hotel: {hotel_url}")
        print("Reason:", e)

    driver.quit()
    return data

```

```python
hotel_urls = [
    "https://www.booking.com/hotel/in/bloomrooms-janpath.en-gb.html",
    "https://www.booking.com/hotel/in/radisson-marina-connaught-place.en-gb.html",
    "https://www.booking.com/hotel/in/uds-villa-new-delhi.en-gb.html",
    "https://www.booking.com/hotel/in/daffodils-new-delhi.en-gb.html",
    "https://www.booking.com/hotel/in/connaught-mews-bed-and-breakfast.en-gb.html"
]
```

```python
all_results = []
for url in hotel_urls:
    results = extract_hotel_data(url)
    all_results.extend(results)

import pandas as pd
df = pd.DataFrame(all_results)
df.to_excel("booking_data.xlsx", index=False)

```
